# 08.Std_Adalinesgd_iris

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
from adalinesgd import AdalineSGD

style.use('seaborn-talk')

if __name__ == '__main__':
    df = pd.read_csv('iris.data', header=None)
    y = df.iloc[0:100,4].values
    y = np.where(y=='Iris-setosa',-1,1)
    X = df.iloc[0:100,[0,2]].values
                     
    X_std = np.copy(X)
    X_std[:,0] = (X[:,0] - X[:,0].mean()) / X[:,0].std()
    X_std[:,1] = (X[:,1] - X[:,1].mean()) / X[:,1].std()
                     
    adal = AdalineSGD(eta=0.01, n_iter=15, random_state=1).fit(X_std, y)
    plt.plot(range(1, len(adal.cost_)+1), adal.cost_, marker='o')
    plt.xlabel('Epochs')
    plt.ylabel('Average Cost')
    plt.title('Adaline Stochastic GD - Learning rate 0.01')
    plt.show()
```

그림에서 보듯이 비용함수가 빠르게 줄어드는 것을 볼 수 있다. 이유는 한번의 반복문에서 가중치를 업데이는 빈도수가 더 많이 일어나기 때문이다.
반복이 많아지면 결국 배치 경사하강법과 비슷해진다.
